```python
import numpy as np
import matplotlib.pyplot as plt
height=np.random.normal(140,20,1000)
plt.figure(dpi=200)
plt.title("Height distribution")
plt.xlabel("height (cm)")
plt.ylabel("Frequency")
plt.hist(height,bins=30,ec="gold")
```

A histogram is generated that plots random height distribution with `mean=140`,`standard deviation=20` using 1000 randomly generated height values

#### Q3 Difference between jupyter and notepad
* Jupyter notebooks provide an interactive environment where you can execute code in cells and output is generated simultaneously. 
* It lets you easily combine Markdown text and executable Python source code on one canvas called a notebook.
* Whereas notepad is a simple text editor that provides plain text environment without any interactivity.

#### Q4 Difference between jupyter and notebook
 * #### Interpreter
   * Translates program one statement at a time.
   * The interpreter converts high-level language to machine language.
   * Eg:Python, Ruby, Perl, SNOBOL, MATLAB, etc
 * #### Compiler
   * For Converting the code written in a high-level language into machine-level language so that computers can easily          understand, we use a compiler
   * Compiled code runs faster in comparison to Interpreted code.
   * Eg:C, C++, C#, etc.
