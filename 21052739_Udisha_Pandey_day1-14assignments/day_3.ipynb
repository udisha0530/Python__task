#### QUES1  What are Data Structures? List some of data structures and their use in real world application? 


Data Structures are a way of organizing data so that it can be accessed more efficiently depending upon the situation. Data Structures are fundamentals of any programming language around which a program is built.

SOME OF THE DATA STRUCTURES AREAS FOLLOWS*:

#### List: 
1.Collection of different python objects.

2.Data is stored in sequential manner.

3.Lists are mutable in nature
##### Application:
   * To store and manipulate collections of data, such as a list of students in a class

#### Tuples:
1.Collection of different python objects.

2.Data stored in Sequential Manner.

3.Immutable in nature.
##### Application:
   * To represent and pass around multiple related values, such as coordinates (latitude, longitude)

#### Dictionaries:
1. Dictionaries are key-value pairs enclosed in curly braces ({}) or created using the dict() constructor. 
2. They provide fast access to values based on their associated keys. 
##### Application:
  *  commonly used for mapping and storing structured data.
#### Sets:
1. ets are unordered collections of unique elements. 
2. They are defined using curly braces ({}) or the set() function.
##### Application: 
  * Sets are useful for operations like membership testing and eliminating duplicates.

###  Ques4 What is difference between ordered data type and unordered data type?  

#### Ordered data Type                                                

1.  Elements are arranged in a specific sequence that can be accessed and iterated in a predictable manner. 
2. Examples: List,Tuples
####  Unordered Data Type
1.   The elements are stored in a way that does not guarantee a specific sequence, and accessing elements based on their position or index is not meaningful.
2. Examples:Sets ,Dictionaries







```python
a=100
print(type(a))
a=105.5
print(type(a))
a=192.56j
print(type(a))
a=10+6j
print(type(a))
a='10'
print(type(a))
a='hello world'
print(type(a))
a=[10,20,50,100]
print(type(a))
a={'name':'Sachin','age':24,'language':'python'}
print(type(a))
```
